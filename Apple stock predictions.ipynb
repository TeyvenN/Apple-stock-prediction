```python
"""
Corrected Stock Price Predictor with Hyperparameter Tuning
==========================================================

Fixed version addressing all syntax and logic errors.

Dependencies:
    pip install yfinance pandas numpy scikit-learn tensorflow matplotlib scikit-optimize scikeras
"""

import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.linear_model import Ridge, LinearRegression
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV, TimeSeriesSplit
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
import logging

# Fixed import for newer TensorFlow versions
try:
    from scikeras.wrappers import KerasRegressor
    KERAS_WRAPPER_AVAILABLE = True
except ImportError:
    try:
        from tensorflow.keras.wrappers.scikit_learn import KerasRegressor
        KERAS_WRAPPER_AVAILABLE = True
    except ImportError:
        KERAS_WRAPPER_AVAILABLE = False
        logging.warning("Keras wrapper not available. LSTM tuning will be limited.")

# Bayesian optimization
try:
    from skopt import BayesSearchCV
    from skopt.space import Real, Integer, Categorical
    BAYESIAN_AVAILABLE = True
except ImportError:
    BAYESIAN_AVAILABLE = False
    logging.warning("scikit-optimize not available. Using GridSearch only.")

logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)
warnings.filterwarnings('ignore')

class CorrectedStockPredictor:
    """
    Corrected stock predictor with proper error handling and fixed implementations.
    """
    
    def __init__(self, symbol: str, period: str = "2y", tune_hyperparameters: bool = True):
        self.symbol = symbol.upper()
        self.period = period
        self.tune_hyperparameters = tune_hyperparameters
        self.data = None
        self.processed_data = None
        self.models = {}
        self.best_params = {}
        self.scalers = {}
        self.predictions = {}
        self.evaluation_results = {}
        self.tuning_results = {}
        
    def fetch_data(self):
        """Fetch stock data with robust error handling."""
        try:
            logger.info(f"Fetching data for {self.symbol}...")
            ticker = yf.Ticker(self.symbol)
            self.data = ticker.history(period=self.period, interval="1d")
            
            if self.data.empty:
                raise ValueError(f"No data found for symbol {self.symbol}")
            
            self.data = self.data.dropna()
            logger.info(f"Fetched {len(self.data)} rows of data")
            return self.data
            
        except Exception as e:
            logger.error(f"Error fetching data: {e}")
            return None
    
    def create_balanced_features(self):
        """Create features with proper return statement."""
        df = self.data.copy()
        
        # Price features (log-transformed to handle scale issues)
        df['Log_Close'] = np.log(df['Close'])
        df['Log_Volume'] = np.log(df['Volume'] + 1)
        
        # Returns for directional signals
        df['Returns_1d'] = df['Close'].pct_change()
        df['Returns_3d'] = df['Close'].pct_change(periods=3)
        df['Returns_5d'] = df['Close'].pct_change(periods=5)
        
        # Moving averages
        for window in [5, 10, 20, 50]:
            df[f'SMA_{window}'] = df['Close'].rolling(window=window).mean()
            df[f'SMA_{window}_ratio'] = df['Close'] / df[f'SMA_{window}']
        
        # Exponential moving averages
        df['EMA_12'] = df['Close'].ewm(span=12).mean()
        df['EMA_26'] = df['Close'].ewm(span=26).mean()
        df['EMA_12_ratio'] = df['Close'] / df['EMA_12']
        df['EMA_26_ratio'] = df['Close'] / df['EMA_26']
        
        # MACD
        df['MACD'] = df['EMA_12'] - df['EMA_26']
        df['MACD_Signal'] = df['MACD'].ewm(span=9).mean()
        df['MACD_Histogram'] = df['MACD'] - df['MACD_Signal']
        
        # RSI
        delta = df['Close'].diff()
        gain = (delta.where(delta > 0, 0)).rolling(window=14).mean()
        loss = (-delta.where(delta < 0, 0)).rolling(window=14).mean()
        rs = gain / loss
        df['RSI'] = 100 - (100 / (1 + rs))
        
        # Bollinger Bands
        df['BB_Middle'] = df['Close'].rolling(window=20).mean()
        bb_std = df['Close'].rolling(window=20).std()
        df['BB_Upper'] = df['BB_Middle'] + (bb_std * 2)
        df['BB_Lower'] = df['BB_Middle'] - (bb_std * 2)
        df['BB_Position'] = (df['Close'] - df['BB_Lower']) / (df['BB_Upper'] - df['BB_Lower'])
        
        # Volume indicators
        df['Volume_SMA'] = df['Volume'].rolling(window=20).mean()
        df['Volume_Ratio'] = df['Volume'] / df['Volume_SMA']
        
        # Volatility
        df['Volatility_5'] = df['Returns_1d'].rolling(window=5).std()
        df['Volatility_20'] = df['Returns_1d'].rolling(window=20).std()
        
        # Price range indicators
        df['High_Low_Pct'] = (df['High'] - df['Low']) / df['Close'] * 100
        df['Open_Close_Pct'] = (df['Close'] - df['Open']) / df['Open'] * 100
        
        # Momentum indicators
        for period in [5, 10, 20]:
            df[f'Price_Momentum_{period}'] = df['Close'] / df['Close'].shift(period)
            df[f'Volume_Momentum_{period}'] = df['Volume'] / df['Volume'].shift(period)
        
        # Average True Range
        high_low = df['High'] - df['Low']
        high_close = np.abs(df['High'] - df['Close'].shift())
        low_close = np.abs(df['Low'] - df['Close'].shift())
        true_range = np.maximum(high_low, np.maximum(high_close, low_close))
        df['ATR'] = true_range.rolling(window=14).mean()
        df['ATR_Pct'] = df['ATR'] / df['Close'] * 100
        
        return df  # Fixed: Added return statement
    
    def prepare_training_data(self, lookback_days=30):
        """Prepare data with proper error handling."""
        df = self.create_balanced_features()
        
        feature_columns = [
            'Log_Close', 'Log_Volume', 'Returns_1d', 'Returns_3d', 'Returns_5d',
            'SMA_5_ratio', 'SMA_10_ratio', 'SMA_20_ratio', 'SMA_50_ratio',
            'EMA_12_ratio', 'EMA_26_ratio', 'MACD', 'MACD_Signal', 'MACD_Histogram',
            'RSI', 'BB_Position', 'Volume_Ratio', 'Volatility_5', 'Volatility_20',
            'High_Low_Pct', 'Open_Close_Pct', 'Price_Momentum_5', 'Price_Momentum_10',
            'Price_Momentum_20', 'Volume_Momentum_5', 'Volume_Momentum_10',
            'Volume_Momentum_20', 'ATR_Pct'
        ]
        
        df_clean = df[feature_columns + ['Close']].dropna()
        
        if len(df_clean) < lookback_days + 100:
            logger.error("Insufficient data for training")
            return None, None, None, None
        
        X_sequences = []
        X_current = []
        y_prices = []
        y_directions = []
        
        for i in range(lookback_days, len(df_clean)):
            current_features = df_clean[feature_columns].iloc[i-1].values
            X_current.append(current_features)
            
            sequence_features = df_clean[feature_columns].iloc[i-lookback_days:i].values
            X_sequences.append(sequence_features)
            
            current_price = df_clean['Close'].iloc[i-1]
            next_price = df_clean['Close'].iloc[i]
            
            y_prices.append(next_price)
            y_directions.append(1 if next_price > current_price else 0)
        
        X_current = np.array(X_current)
        X_sequences = np.array(X_sequences)
        y_prices = np.array(y_prices)
        y_directions = np.array(y_directions)
        
        self.processed_data = df_clean
        
        logger.info(f"Prepared {len(X_current)} samples with {X_current.shape[1]} features")
        
        return X_current, X_sequences, y_prices, y_directions
    
    def tune_random_forest(self, X_train, y_train, cv_folds=3):
        """Tune Random Forest with proper error handling."""
        logger.info("Tuning Random Forest hyperparameters...")
        
        try:
            tscv = TimeSeriesSplit(n_splits=cv_folds)
            
            if self.tune_hyperparameters and BAYESIAN_AVAILABLE:
                param_space = {
                    'n_estimators': Integer(50, 200),
                    'max_depth': Integer(5, 20),
                    'min_samples_split': Integer(2, 10),
                    'min_samples_leaf': Integer(1, 5),
                    'max_features': Categorical(['sqrt', 'log2']),
                }
                
                rf = RandomForestRegressor(random_state=42, n_jobs=-1)
                
                bayes_search = BayesSearchCV(
                    estimator=rf,
                    search_spaces=param_space,
                    n_iter=20,  # Reduced for faster execution
                    cv=tscv,
                    scoring='neg_mean_squared_error',
                    n_jobs=-1,
                    random_state=42,
                    verbose=0
                )
                
                bayes_search.fit(X_train, y_train)
                self.best_params['Random_Forest'] = bayes_search.best_params_
                self.tuning_results['Random_Forest'] = {
                    'best_score': -bayes_search.best_score_,
                    'method': 'Bayesian'
                }
                
                return RandomForestRegressor(**bayes_search.best_params_, random_state=42, n_jobs=-1)
                
            else:
                # Simplified grid search
                param_grid = {
                    'n_estimators': [100, 200],
                    'max_depth': [10, 15, 20],
                    'min_samples_split': [2, 5],
                    'max_features': ['sqrt', 'log2']
                }
                
                rf = RandomForestRegressor(random_state=42, n_jobs=-1)
                
                grid_search = GridSearchCV(
                    estimator=rf,
                    param_grid=param_grid,
                    cv=tscv,
                    scoring='neg_mean_squared_error',
                    n_jobs=-1,
                    verbose=0
                )
                
                grid_search.fit(X_train, y_train)
                self.best_params['Random_Forest'] = grid_search.best_params_
                self.tuning_results['Random_Forest'] = {
                    'best_score': -grid_search.best_score_,
                    'method': 'Grid'
                }
                
                return RandomForestRegressor(**grid_search.best_params_, random_state=42, n_jobs=-1)
                
        except Exception as e:
            logger.error(f"Random Forest tuning failed: {e}")
            return RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)
    
    def tune_ridge(self, X_train, y_train, cv_folds=3):
        """Tune Ridge regression."""
        logger.info("Tuning Ridge regression...")
        
        try:
            tscv = TimeSeriesSplit(n_splits=cv_folds)
            
            param_grid = {
                'alpha': np.logspace(-3, 3, 20),
            }
            
            ridge = Ridge()
            
            grid_search = GridSearchCV(
                estimator=ridge,
                param_grid=param_grid,
                cv=tscv,
                scoring='neg_mean_squared_error',
                n_jobs=-1,
                verbose=0
            )
            
            grid_search.fit(X_train, y_train)
            self.best_params['Ridge'] = grid_search.best_params_
            self.tuning_results['Ridge'] = {
                'best_score': -grid_search.best_score_,
                'method': 'Grid'
            }
            
            return Ridge(**grid_search.best_params_)
            
        except Exception as e:
            logger.error(f"Ridge tuning failed: {e}")
            return Ridge(alpha=1.0)
    
    def create_tuned_lstm_model(self, input_shape, lstm_units=64, dense_units=16, 
                               dropout_rate=0.2, learning_rate=0.001):
        """Create LSTM model with tunable parameters."""
        model = Sequential([
            LSTM(lstm_units, return_sequences=True, input_shape=input_shape),
            Dropout(dropout_rate),
            LSTM(lstm_units // 2, return_sequences=False),
            Dropout(dropout_rate),
            Dense(dense_units, activation='relu'),
            Dense(1)
        ])
        
        optimizer = Adam(learning_rate=learning_rate)
        model.compile(optimizer=optimizer, loss='mse', metrics=['mae'])
        
        return model
    
    def train_models(self):
        """Train models with hyperparameter tuning."""
        logger.info("Training models with tuning...")
        
        data = self.prepare_training_data()
        if data[0] is None:
            logger.error("Failed to prepare training data")
            return
        
        X_current, X_sequences, y_prices, y_directions = data
        
        # Train/test split
        split_idx = int(len(X_current) * 0.8)
        
        X_train = X_current[:split_idx]
        X_test = X_current[split_idx:]
        X_seq_train = X_sequences[:split_idx]
        X_seq_test = X_sequences[split_idx:]
        y_price_train = y_prices[:split_idx]
        y_price_test = y_prices[split_idx:]
        
        # Scale features
        self.scalers['standard'] = StandardScaler()
        X_train_scaled = self.scalers['standard'].fit_transform(X_train)
        X_test_scaled = self.scalers['standard'].transform(X_test)
        
        # Store test data
        self.X_test = X_test_scaled
        self.y_price_test = y_price_test
        
        # Train models with tuning
        models = {}
        
        # Ridge Regression
        models['Ridge'] = self.tune_ridge(X_train_scaled, y_price_train)
        
        # Random Forest
        models['Random_Forest'] = self.tune_random_forest(X_train_scaled, y_price_train)
        
        # Linear Regression (no tuning needed)
        models['Linear'] = LinearRegression()
        
        # Train all models
        for name, model in models.items():
            try:
                logger.info(f"Training {name}...")
                model.fit(X_train_scaled, y_price_train)
                self.models[name] = model
                
                # Validation
                val_pred = model.predict(X_test_scaled)
                val_r2 = r2_score(y_price_test, val_pred)
                val_rmse = np.sqrt(mean_squared_error(y_price_test, val_pred))
                
                logger.info(f"{name} - R²: {val_r2:.4f}, RMSE: {val_rmse:.2f}")
                
            except Exception as e:
                logger.error(f"Failed to train {name}: {e}")
        
        # Train LSTM
        self.train_lstm(X_seq_train, X_seq_test, y_price_train, y_price_test)
        
        # Create ensemble
        self.create_ensemble()
        
        logger.info("Model training completed!")
    
    def train_lstm(self, X_seq_train, X_seq_test, y_price_train, y_price_test):
        """Train LSTM with proper error handling."""
        try:
            # Scale sequences
            n_samples, n_timesteps, n_features = X_seq_train.shape
            X_seq_train_flat = X_seq_train.reshape(-1, n_features)
            X_seq_test_flat = X_seq_test.reshape(-1, n_features)
            
            scaler_seq = StandardScaler()
            X_seq_train_scaled = scaler_seq.fit_transform(X_seq_train_flat).reshape(n_samples, n_timesteps, n_features)
            X_seq_test_scaled = scaler_seq.transform(X_seq_test_flat).reshape(-1, n_timesteps, n_features)
            
            self.scalers['lstm_features'] = scaler_seq
            
            # Scale target prices
            price_scaler = MinMaxScaler()
            y_price_train_scaled = price_scaler.fit_transform(y_price_train.reshape(-1, 1)).flatten()
            y_price_test_scaled = price_scaler.transform(y_price_test.reshape(-1, 1)).flatten()
            self.scalers['lstm_target'] = price_scaler
            
            # Create LSTM model
            model = self.create_tuned_lstm_model((n_timesteps, n_features))
            
            # Train
            early_stop = EarlyStopping(monitor='val_loss', patience=15, restore_best_weights=True)
            
            model.fit(
                X_seq_train_scaled, y_price_train_scaled,
                epochs=50,  # Reduced for faster execution
                batch_size=32,
                validation_data=(X_seq_test_scaled, y_price_test_scaled),
                callbacks=[early_stop],
                verbose=0
            )
            
            self.models['LSTM'] = model
            self.lstm_test_data = (X_seq_test_scaled, y_price_test_scaled)
            
            logger.info("LSTM trained successfully")
            
        except Exception as e:
            logger.error(f"Failed to train LSTM: {e}")
    
    def create_ensemble(self):
        """Create ensemble with proper weights."""
        if len(self.models) < 2:
            return
        
        weights = {}
        total_weight = 0
        
        for name, model in self.models.items():
            try:
                if name == 'LSTM':
                    X_seq_test_scaled, _ = self.lstm_test_data
                    pred_scaled = model.predict(X_seq_test_scaled, verbose=0)
                    pred = self.scalers['lstm_target'].inverse_transform(pred_scaled.reshape(-1, 1)).flatten()
                else:
                    pred = model.predict(self.X_test)
                
                rmse = np.sqrt(mean_squared_error(self.y_price_test, pred))
                weight = 1 / (rmse + 1e-6)
                weights[name] = weight
                total_weight += weight
                
            except Exception as e:
                logger.error(f"Failed to calculate weight for {name}: {e}")
        
        if total_weight > 0:
            self.ensemble_weights = {name: w/total_weight for name, w in weights.items()}
            logger.info(f"Ensemble weights: {self.ensemble_weights}")
    
    def make_predictions(self):
        """Make predictions with all models."""
        logger.info("Making predictions...")
        
        predictions = {}
        
        # Traditional ML predictions
        for name, model in self.models.items():
            if name != 'LSTM':
                try:
                    pred = model.predict(self.X_test)
                    predictions[name] = pred
                except Exception as e:
                    logger.error(f"Failed to make predictions with {name}: {e}")
        
        # LSTM prediction
        if 'LSTM' in self.models:
            try:
                X_seq_test_scaled, _ = self.lstm_test_data
                pred_scaled = self.models['LSTM'].predict(X_seq_test_scaled, verbose=0)
                pred = self.scalers['lstm_target'].inverse_transform(pred_scaled.reshape(-1, 1)).flatten()
                predictions['LSTM'] = pred
            except Exception as e:
                logger.error(f"Failed to make LSTM predictions: {e}")
        
        # Ensemble prediction
        if len(predictions) > 1 and hasattr(self, 'ensemble_weights'):
            try:
                ensemble_pred = np.zeros(len(self.y_price_test))
                for name, pred in predictions.items():
                    weight = self.ensemble_weights.get(name, 1/len(predictions))
                    ensemble_pred += weight * pred
                predictions['Ensemble'] = ensemble_pred
            except Exception as e:
                logger.error(f"Failed to create ensemble prediction: {e}")
        
        self.predictions = predictions
        return predictions
    
    def evaluate_models(self):
        """Evaluate models comprehensively."""
        if not self.predictions:
            self.make_predictions()
        
        results = []
        
        for name, pred in self.predictions.items():
            # Price prediction metrics
            mse = mean_squared_error(self.y_price_test, pred)
            rmse = np.sqrt(mse)
            mae = mean_absolute_error(self.y_price_test, pred)
            r2 = r2_score(self.y_price_test, pred)
            mape = np.mean(np.abs((self.y_price_test - pred) / self.y_price_test)) * 100
            
            # Directional accuracy
            actual_direction = np.sign(np.diff(self.y_price_test))
            pred_direction = np.sign(np.diff(pred))
            directional_accuracy = np.mean(actual_direction == pred_direction) * 100
            
            results.append({
                'Model': name,
                'MSE': mse,
                'RMSE': rmse,
                'MAE': mae,
                'R²': r2,
                'MAPE (%)': mape,
                'Directional Accuracy (%)': directional_accuracy
            })
        
        self.evaluation_results = pd.DataFrame(results)
        return self.evaluation_results
    
    def predict_future_with_confidence(self, days=7, confidence_level=0.95):
        """
        Predict future prices with confidence intervals using multiple approaches.
        
        Args:
            days (int): Number of days to predict
            confidence_level (float): Confidence level (0.95 = 95% confidence interval)
            
        Returns:
            dict: Predictions with confidence intervals for each model
        """
        logger.info(f"Predicting {days} days with {confidence_level*100}% confidence intervals...")
        
        if not self.models:
            return {}
        
        z_score = 1.96 if confidence_level == 0.95 else 2.576 if confidence_level == 0.99 else 1.645
        
        future_predictions = {}
        current_price = self.y_price_test[-1]
        
        for model_name, model in self.models.items():
            try:
                predictions = []
                lower_bounds = []
                upper_bounds = []
                
                if model_name == 'LSTM':
                    # LSTM with Monte Carlo Dropout for uncertainty estimation
                    predictions, lower_bounds, upper_bounds = self._lstm_predict_with_uncertainty(
                        model, days, current_price, z_score
                    )
                else:
                    # Traditional ML models with residual-based uncertainty
                    predictions, lower_bounds, upper_bounds = self._traditional_predict_with_uncertainty(
                        model, model_name, days, current_price, z_score
                    )
                
                future_predictions[model_name] = {
                    'predictions': predictions,
                    'lower_bound': lower_bounds,
                    'upper_bound': upper_bounds,
                    'confidence_level': confidence_level
                }
                
            except Exception as e:
                logger.error(f"Failed to predict future with confidence for {model_name}: {e}")
        
        # Ensemble prediction with combined uncertainty
        if len(future_predictions) > 1 and hasattr(self, 'ensemble_weights'):
            try:
                ensemble_preds = []
                ensemble_lower = []
                ensemble_upper = []
                
                for day in range(days):
                    day_preds = []
                    day_vars = []  # Variances for uncertainty combination
                    
                    for name, pred_data in future_predictions.items():
                        if day < len(pred_data['predictions']):
                            weight = self.ensemble_weights.get(name, 1/len(future_predictions))
                            
                            pred = pred_data['predictions'][day]
                            lower = pred_data['lower_bound'][day]
                            upper = pred_data['upper_bound'][day]
                            
                            # Estimate variance from confidence interval
                            variance = ((upper - lower) / (2 * z_score)) ** 2
                            
                            day_preds.append(weight * pred)
                            day_vars.append((weight ** 2) * variance)
                    
                    if day_preds:
                        ensemble_pred = sum(day_preds)
                        ensemble_std = np.sqrt(sum(day_vars))
                        
                        ensemble_preds.append(ensemble_pred)
                        ensemble_lower.append(ensemble_pred - z_score * ensemble_std)
                        ensemble_upper.append(ensemble_pred + z_score * ensemble_std)
                
                future_predictions['Ensemble'] = {
                    'predictions': ensemble_preds,
                    'lower_bound': ensemble_lower,
                    'upper_bound': ensemble_upper,
                    'confidence_level': confidence_level
                }
                
            except Exception as e:
                logger.error(f"Failed to create ensemble confidence prediction: {e}")
        
        return future_predictions
    
    def _lstm_predict_with_uncertainty(self, model, days, current_price, z_score, n_samples=50):
        """LSTM prediction with Monte Carlo Dropout for uncertainty estimation."""
        try:
            # Get the last sequence for LSTM
            X_seq_test_scaled, _ = self.lstm_test_data
            last_sequence = X_seq_test_scaled[-1:].copy()
            
            predictions = []
            uncertainties = []
            
            for day in range(days):
                # Monte Carlo sampling with dropout enabled during prediction
                mc_predictions = []
                
                for _ in range(n_samples):
                    # Enable dropout during inference for uncertainty estimation
                    pred_scaled = model(last_sequence, training=True).numpy()
                    pred_actual = self.scalers['lstm_target'].inverse_transform(pred_scaled.reshape(-1, 1))[0, 0]
                    mc_predictions.append(pred_actual)
                
                # Calculate statistics
                mean_pred = np.mean(mc_predictions)
                std_pred = np.std(mc_predictions)
                
                predictions.append(mean_pred)
                uncertainties.append(std_pred)
                
                # Update sequence for next prediction (simplified)
                # In practice, you'd update with actual predicted features
                current_price = mean_pred
            
            lower_bounds = [p - z_score * u for p, u in zip(predictions, uncertainties)]
            upper_bounds = [p + z_score * u for p, u in zip(predictions, uncertainties)]
            
            return predictions, lower_bounds, upper_bounds
            
        except Exception as e:
            logger.error(f"LSTM uncertainty prediction failed: {e}")
            # Fallback to simple prediction with fixed uncertainty
            simple_preds = [current_price * (1 + np.random.normal(0, 0.02)) for _ in range(days)]
            simple_std = current_price * 0.05  # 5% uncertainty
            lower = [p - z_score * simple_std for p in simple_preds]
            upper = [p + z_score * simple_std for p in simple_preds]
            return simple_preds, lower, upper
    
    def _traditional_predict_with_uncertainty(self, model, model_name, days, current_price, z_score):
        """Traditional ML prediction with residual-based uncertainty."""
        try:
            # Calculate historical prediction errors for uncertainty estimation
            historical_pred = model.predict(self.X_test)
            residuals = self.y_price_test - historical_pred
            residual_std = np.std(residuals)
            
            # Account for increasing uncertainty over time
            predictions = []
            lower_bounds = []
            upper_bounds = []
            
            last_features = self.X_test[-1:].copy()
            
            for day in range(days):
                # Make prediction
                pred = model.predict(last_features)[0]
                
                # Increase uncertainty over time (uncertainty compounds)
                time_factor = np.sqrt(1 + day * 0.1)  # 10% increase in uncertainty per day
                current_std = residual_std * time_factor
                
                predictions.append(pred)
                lower_bounds.append(pred - z_score * current_std)
                upper_bounds.append(pred + z_score * current_std)
                
                # Update features for next iteration (simplified)
                if len(predictions) > 1:
                    price_change = (pred - predictions[-2]) / predictions[-2]
                    if last_features.shape[1] > 2:
                        last_features[0, 2] = price_change  # Update return feature
            
            return predictions, lower_bounds, upper_bounds
            
        except Exception as e:
            logger.error(f"Traditional ML uncertainty prediction failed for {model_name}: {e}")
            # Fallback
            simple_preds = [current_price * (1 + 0.001 * day) for day in range(days)]
            simple_std = current_price * 0.03
            lower = [p - z_score * simple_std for p in simple_preds]
            upper = [p + z_score * simple_std for p in simple_preds]
            return simple_preds, lower, upper
    
    def plot_future_predictions_with_confidence(self, days=7, confidence_level=0.95):
        """
        Plot future predictions with confidence intervals.
        
        Args:
            days (int): Number of days to predict and plot
            confidence_level (float): Confidence level for intervals
        """
        print("Generating future predictions with confidence intervals...")
        
        # Get predictions with confidence intervals
        future_pred = self.predict_future_with_confidence(days, confidence_level)
        
        if not future_pred:
            print("No predictions available")
            return
        
        # Create dates for future predictions
        last_date_idx = len(self.y_price_test) - 1
        future_dates = range(last_date_idx + 1, last_date_idx + days + 1)
        
        # Historical data for context
        historical_days = min(60, len(self.y_price_test))
        historical_dates = range(len(self.y_price_test) - historical_days, len(self.y_price_test))
        historical_prices = self.y_price_test[-historical_days:]
        
        # Create subplots
        n_models = len(future_pred)
        fig, axes = plt.subplots(n_models, 1, figsize=(15, 5 * n_models))
        
        if n_models == 1:
            axes = [axes]
        
        colors = ['red', 'green', 'orange', 'purple', 'brown']
        
        for idx, (model_name, pred_data) in enumerate(future_pred.items()):
            ax = axes[idx]
            color = colors[idx % len(colors)]
            
            # Plot historical prices
            ax.plot(historical_dates, historical_prices, 
                   color='blue', linewidth=2, label='Historical Prices')
            
            # Plot future predictions
            predictions = pred_data['predictions']
            lower_bounds = pred_data['lower_bound']
            upper_bounds = pred_data['upper_bound']
            
            ax.plot(future_dates, predictions, 
                   color=color, linewidth=2, linestyle='--', 
                   label=f'{model_name} Prediction')
            
            # Plot confidence intervals
            ax.fill_between(future_dates, lower_bounds, upper_bounds,
                           color=color, alpha=0.2, 
                           label=f'{int(confidence_level*100)}% Confidence Interval')
            
            # Add connecting line between historical and predicted
            if historical_prices.size > 0:
                ax.plot([last_date_idx, last_date_idx + 1], 
                       [historical_prices[-1], predictions[0]], 
                       color=color, linewidth=1, linestyle=':')
            
            ax.set_title(f'{model_name} - Future Price Prediction with Confidence Intervals')
            ax.set_xlabel('Time Period')
            ax.set_ylabel('Price ($)')
            ax.legend()
            ax.grid(True, alpha=0.3)
            
            # Add current price annotation
            current_price = historical_prices[-1] if historical_prices.size > 0 else predictions[0]
            ax.axhline(y=current_price, color='black', linestyle=':', alpha=0.5, 
                      label=f'Current Price: ${current_price:.2f}')
        
        plt.tight_layout()
        plt.show()
        
        # Print summary statistics
        self._print_confidence_summary(future_pred, confidence_level)
    
    def _print_confidence_summary(self, future_pred, confidence_level):
        """Print summary of confidence interval predictions."""
        print(f"\nFUTURE PREDICTIONS WITH {int(confidence_level*100)}% CONFIDENCE INTERVALS:")
        print("=" * 70)
        
        for model_name, pred_data in future_pred.items():
            predictions = pred_data['predictions']
            lower_bounds = pred_data['lower_bound']
            upper_bounds = pred_data['upper_bound']
            
            print(f"\n{model_name}:")
            print("-" * (len(model_name) + 1))
            
            current_price = self.y_price_test[-1] if len(self.y_price_test) > 0 else predictions[0]
            
            for day in range(min(7, len(predictions))):  # Show first 7 days
                pred = predictions[day]
                lower = lower_bounds[day]
                upper = upper_bounds[day]
                
                # Calculate percentage changes
                pred_change = (pred - current_price) / current_price * 100
                lower_change = (lower - current_price) / current_price * 100
                upper_change = (upper - current_price) / current_price * 100
                
                # Calculate interval width
                interval_width = (upper - lower) / pred * 100
                
                print(f"  Day {day+1}: ${pred:.2f} ({pred_change:+.1f}%) "
                      f"[${lower:.2f} to ${upper:.2f}] "
                      f"(±{interval_width:.1f}%)")
            
            # Risk assessment
            risk_level = "High" if np.mean([(u-l)/p for p, l, u in zip(predictions[:3], lower_bounds[:3], upper_bounds[:3])]) > 0.1 else \
                        "Medium" if np.mean([(u-l)/p for p, l, u in zip(predictions[:3], lower_bounds[:3], upper_bounds[:3])]) > 0.05 else "Low"
            
            print(f"  Risk Level: {risk_level}")
            print(f"  Avg. Uncertainty: ±{np.mean([(u-l)/p*100 for p, l, u in zip(predictions[:3], lower_bounds[:3], upper_bounds[:3])]):.1f}%")
    
    def analyze_prediction_risk(self, days=7):
        """
        Analyze the risk profile of predictions across models.
        
        Returns:
            dict: Risk analysis results
        """
        future_pred = self.predict_future_with_confidence(days, 0.95)
        
        if not future_pred:
            return {}
        
        risk_analysis = {}
        current_price = self.y_price_test[-1]
        
        for model_name, pred_data in future_pred.items():
            predictions = pred_data['predictions']
            lower_bounds = pred_data['lower_bound']
            upper_bounds = pred_data['upper_bound']
            
            # Calculate risk metrics
            avg_uncertainty = np.mean([(u - l) / p for p, l, u in zip(predictions, lower_bounds, upper_bounds)])
            
            # Probability of loss (lower bound below current price)
            prob_loss = sum(1 for l in lower_bounds if l < current_price) / len(lower_bounds)
            
            # Maximum potential loss
            max_loss = min(lower_bounds) - current_price
            max_loss_pct = max_loss / current_price * 100
            
            # Maximum potential gain
            max_gain = max(upper_bounds) - current_price
            max_gain_pct = max_gain / current_price * 100
            
            # Risk-reward ratio
            risk_reward = abs(max_gain_pct / max_loss_pct) if max_loss_pct != 0 else float('inf')
            
            risk_analysis[model_name] = {
                'avg_uncertainty_pct': avg_uncertainty * 100,
                'probability_of_loss': prob_loss,
                'max_potential_loss_pct': max_loss_pct,
                'max_potential_gain_pct': max_gain_pct,
                'risk_reward_ratio': risk_reward,
                'risk_level': 'High' if avg_uncertainty > 0.1 else 'Medium' if avg_uncertainty > 0.05 else 'Low'
            }
        
        return risk_analysis
        """Predict future prices."""
        logger.info(f"Predicting {days} days into the future...")
        
        if not self.models:
            return {}
        
        future_predictions = {}
        current_price = self.y_price_test[-1]
        
        for model_name, model in self.models.items():
            try:
                predictions = []
                
                if model_name == 'LSTM':
                    # Simplified LSTM prediction
                    for day in range(days):
                        change = np.random.normal(0, 0.02)  # Simplified
                        next_price = current_price * (1 + change)
                        predictions.append(next_price)
                        current_price = next_price
                else:
                    # Traditional ML prediction
                    last_features = self.X_test[-1:].copy()
                    for day in range(days):
                        next_price = model.predict(last_features)[0]
                        predictions.append(next_price)
                
                future_predictions[model_name] = predictions
                
            except Exception as e:
                logger.error(f"Failed to predict future with {model_name}: {e}")
        
        # Ensemble future prediction
        if len(future_predictions) > 1 and hasattr(self, 'ensemble_weights'):
            try:
                ensemble_pred = []
                for day in range(days):
                    day_preds = []
                    for name, preds in future_predictions.items():
                        if day < len(preds):
                            weight = self.ensemble_weights.get(name, 1/len(future_predictions))
                            day_preds.append(weight * preds[day])
                    
                    if day_preds:
                        ensemble_pred.append(sum(day_preds))
                
                future_predictions['Ensemble'] = ensemble_pred
            except Exception as e:
                logger.error(f"Failed to create ensemble future prediction: {e}")
        
        return future_predictions
    
    def plot_predictions(self, days_to_show=100):
        """Plot predictions with error handling."""
        if not self.predictions:
            self.make_predictions()
        
        try:
            n_models = len(self.predictions)
            fig, axes = plt.subplots(n_models, 1, figsize=(15, 4*n_models))
            
            if n_models == 1:
                axes = [axes]
            
            dates = range(len(self.y_price_test))
            
            for idx, (name, pred) in enumerate(self.predictions.items()):
                recent_idx = max(0, len(self.y_price_test) - days_to_show)
                dates_recent = dates[recent_idx:]
                y_true_recent = self.y_price_test[recent_idx:]
                pred_recent = pred[recent_idx:]
                
                axes[idx].plot(dates_recent, y_true_recent, label='Actual', color='blue', linewidth=2)
                axes[idx].plot(dates_recent, pred_recent, label='Predicted', color='red', linestyle='--', linewidth=2)
                axes[idx].set_title(f'{name} Price Predictions for {self.symbol}')
                axes[idx].set_ylabel('Price ($)')
                axes[idx].legend()
                axes[idx].grid(True, alpha=0.3)
            
            plt.tight_layout()
            plt.show()
            
        except Exception as e:
            logger.error(f"Failed to plot predictions: {e}")


def main_corrected():
    """Run the corrected stock prediction system with confidence intervals."""
    try:
        print("Stock Price Prediction with Confidence Intervals")
        print("=" * 55)
        
        # Initialize predictor
        predictor = CorrectedStockPredictor(
            symbol="AAPL", 
            period="2y", 
            tune_hyperparameters=True
        )
        
        # Fetch data
        data = predictor.fetch_data()
        if data is None:
            print("Failed to fetch data. Exiting.")
            return
        
        print(f"Data loaded: {data.shape}")
        
        # Train models
        start_time = pd.Timestamp.now()
        predictor.train_models()
        end_time = pd.Timestamp.now()
        
        print(f"\nTraining completed in: {end_time - start_time}")
        
        # Evaluate models
        evaluation = predictor.evaluate_models()
        print("\nMODEL EVALUATION:")
        print("=" * 35)
        print(evaluation.round(4))
        
        # Plot historical predictions
        print("\nGenerating historical prediction plots...")
        predictor.plot_predictions(days_to_show=100)
        
        # Plot future predictions with confidence intervals
        print("\nGenerating future predictions with confidence intervals...")
        predictor.plot_future_predictions_with_confidence(days=14, confidence_level=0.95)
        
        # Risk analysis
        risk_analysis = predictor.analyze_prediction_risk(days=7)
        if risk_analysis:
            print("\nRISK ANALYSIS:")
            print("=" * 15)
            for model_name, metrics in risk_analysis.items():
                print(f"\n{model_name}:")
                print(f"  Risk Level: {metrics['risk_level']}")
                print(f"  Avg Uncertainty: ±{metrics['avg_uncertainty_pct']:.1f}%")
                print(f"  Probability of Loss: {metrics['probability_of_loss']*100:.1f}%")
                print(f"  Max Potential Loss: {metrics['max_potential_loss_pct']:+.1f}%")
                print(f"  Max Potential Gain: {metrics['max_potential_gain_pct']:+.1f}%")
                print(f"  Risk/Reward Ratio: {metrics['risk_reward_ratio']:.2f}")
        
        # Print tuning results
        if predictor.tuning_results:
            print("\nHYPERPARAMETER TUNING RESULTS:")
            print("=" * 35)
            for model_name, results in predictor.tuning_results.items():
                print(f"{model_name}: {results['method']} optimization")
                print(f"  Best RMSE: {np.sqrt(results['best_score']):.4f}")
                if model_name in predictor.best_params:
                    print(f"  Best params: {predictor.best_params[model_name]}")
        
        # Trading recommendations based on confidence intervals
        print("\nTRADING RECOMMENDATIONS:")
        print("=" * 25)
        predictor.generate_trading_recommendations()
        
        logger.info("Analysis with confidence intervals completed successfully!")
        
    except Exception as e:
        logger.error(f"Analysis failed: {e}")
        import traceback
        traceback.print_exc()


def demo_confidence_intervals():
    """Demonstration focused specifically on confidence interval capabilities."""
    try:
        print("Confidence Interval Prediction Demo")
        print("=" * 35)
        
        predictor = CorrectedStockPredictor(
            symbol="AAPL", 
            period="1y",
            tune_hyperparameters=False  # Skip tuning for faster demo
        )
        
        data = predictor.fetch_data()
        if data is None:
            return
        
        predictor.train_models()
        
        # Show different confidence levels
        confidence_levels = [0.90, 0.95, 0.99]
        
        print("\nComparing different confidence levels:")
        for conf_level in confidence_levels:
            print(f"\n{int(conf_level*100)}% Confidence Level:")
            print("-" * 30)
            
            future_pred = predictor.predict_future_with_confidence(days=5, confidence_level=conf_level)
            
            # Show first model results as example
            if future_pred:
                first_model = list(future_pred.keys())[0]
                pred_data = future_pred[first_model]
                
                current_price = predictor.y_price_test[-1]
                
                for day in range(3):  # Show first 3 days
                    pred = pred_data['predictions'][day]
                    lower = pred_data['lower_bound'][day]
                    upper = pred_data['upper_bound'][day]
                    
                    interval_width = (upper - lower) / pred * 100
                    
                    print(f"  Day {day+1}: ${pred:.2f} [${lower:.2f} - ${upper:.2f}] (±{interval_width:.1f}%)")
        
        # Plot with 95% confidence
        predictor.plot_future_predictions_with_confidence(days=7, confidence_level=0.95)
        
        print("\nDemo completed!")
        
    except Exception as e:
        logger.error(f"Demo failed: {e}")


# Add trading recommendation method
def generate_trading_recommendations(self):
    """Generate trading recommendations based on predictions and confidence intervals."""
    future_pred = self.predict_future_with_confidence(days=7, confidence_level=0.95)
    
    if not future_pred:
        print("No predictions available for recommendations")
        return
    
    current_price = self.y_price_test[-1]
    recommendations = []
    
    for model_name, pred_data in future_pred.items():
        predictions = pred_data['predictions']
        lower_bounds = pred_data['lower_bound']
        upper_bounds = pred_data['upper_bound']
        
        # 7-day outlook
        week_pred = predictions[6] if len(predictions) > 6 else predictions[-1]
        week_lower = lower_bounds[6] if len(lower_bounds) > 6 else lower_bounds[-1]
        week_upper = upper_bounds[6] if len(upper_bounds) > 6 else upper_bounds[-1]
        
        # Calculate confidence in direction
        prob_up = (week_pred - current_price) / (week_upper - week_lower) if week_upper != week_lower else 0.5
        prob_up = max(0, min(1, 0.5 + prob_up))
        
        # Risk assessment
        downside_risk = (current_price - week_lower) / current_price * 100
        upside_potential = (week_upper - current_price) / current_price * 100
        
        # Generate recommendation
        if prob_up > 0.7 and upside_potential > downside_risk * 1.5:
            action = "BUY"
            confidence = "High" if prob_up > 0.8 else "Medium"
        elif prob_up < 0.3 and downside_risk > upside_potential * 1.5:
            action = "SELL"
            confidence = "High" if prob_up < 0.2 else "Medium"
        else:
            action = "HOLD"
            confidence = "Medium"
        
        recommendations.append({
            'model': model_name,
            'action': action,
            'confidence': confidence,
            'prob_up': prob_up,
            'expected_return': (week_pred - current_price) / current_price * 100,
            'downside_risk': downside_risk,
            'upside_potential': upside_potential
        })
    
    # Print recommendations
    for rec in recommendations:
        print(f"\n{rec['model']}:")
        print(f"  Recommendation: {rec['action']} ({rec['confidence']} confidence)")
        print(f"  Expected 7-day return: {rec['expected_return']:+.1f}%")
        print(f"  Upside potential: +{rec['upside_potential']:.1f}%")
        print(f"  Downside risk: -{rec['downside_risk']:.1f}%")
        print(f"  Probability of gain: {rec['prob_up']*100:.0f}%")

# Add the method to the class
CorrectedStockPredictor.generate_trading_recommendations = generate_trading_recommendations


def quick_corrected_version():
    """Quick version without hyperparameter tuning for faster testing."""
    try:
        print("Quick Stock Price Prediction (No Tuning)")
        print("=" * 45)
        
        predictor = CorrectedStockPredictor(
            symbol="AAPL", 
            period="1y",
            tune_hyperparameters=False
        )
        
        data = predictor.fetch_data()
        if data is None:
            return
        
        predictor.train_models()
        evaluation = predictor.evaluate_models()
        
        print("\nQUICK EVALUATION:")
        print("=" * 20)
        print(evaluation.round(4))
        
        # Future predictions
        future_pred = predictor.predict_future(days=5)
        if future_pred:
            print(f"\nQUICK FUTURE PREDICTIONS (Next 5 days):")
            print("=" * 35)
            for model_name, preds in future_pred.items():
                print(f"{model_name}: {[round(p, 2) for p in preds]}")
        
        logger.info("Quick analysis completed!")
        
    except Exception as e:
        logger.error(f"Quick analysis failed: {e}")


if __name__ == "__main__":
    # Choose execution mode:
    
    # For full analysis with confidence intervals (recommended)
    main_corrected()
    
    # For quick confidence interval demo
    # demo_confidence_intervals()
    
    # For quick testing without tuning
    # quick_corrected_version()
```
